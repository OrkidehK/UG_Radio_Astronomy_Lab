```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import ugradio 
import csv
```

```python

coldFile = "hornCold_1419MHzLO_signalRF_maxSamp"
humanFile = "hornHUMAN_1421MHzLO_signalRF_maxSamp"


#opening files and reorganizing them ##
dataCold= [] 
with open(coldFile, 'r') as fc:
    strs = [s.split(')') for s in fc]
    for i in range(len(strs)):
            dataCold.append([eval(num.strip('( ')) for num in strs[i][:-1]])

dataHuman =[]
with open(humanFile, 'r') as fh:
    strs = [s.split(')') for s in fh]
    for i in range(len(strs)):
        dataHuman.append([eval(num.strip('( ')) for num in strs[i][:-1]])

        
#once we have the data do: 

deltaT = 10 #T_cold - T_human = 10 Kelvin
voltHuman = np.fft.fft(dataHuman)
voltCold = np.fft.fft(dataCold)

pwrHuman = np.mean(np.real(voltHuman)**2 + np.imag(voltHuman)**2)
pwrCold = np.mean(np.real(voltCold)**2 + np.imag(voltCold)**2)

gain = (deltaT * np.array(pwrCold))/ (np.array(pwrHuman) - np.array(pwrCold))

freqs = np.fft.fftfreq(2048, 1/3.2e6)

power = np.mean(np.fft.fftshift(np.abs(np.fft.fft(dataCold)))**2, axis = 0)
powerH = np.mean(np.fft.fftshift(np.abs(np.fft.fft(dataHuman)))**2, axis = 0)

fig, ax = plt.subplots()
# ax.semilogy(np.fft.fftshift(freqs), np.fft.fftshift(pwrHuman))
# ax.semilogy(np.fft.fftshift(freqs), np.fft.fftshift(pwrCold))
ax.semilogy(np.fft.fftshift(freqs), power, label ='cold sky', color ='b')
ax.semilogy(np.fft.fftshift(freqs), powerH, label = 'convinient blackbody', color ='r')
plt.legend()
plt.grid()
plt.title('power spectra of cold sky vs people')
plt.xlabel('frequency [MHz]')
plt.ylabel('Power [$V^2$]')
plt.show()


print(len(dataCold))
print(np.shape(dataCold))

print(np.mean(gain))

```

```python
coldFile = "hornCOLD_1420_906MHzLO_signalRF_maxSamp"
dataCold= [] 
with open(coldFile, 'r') as fc:
    strs = [s.split(')') for s in fc]
    for i in range(len(strs)):
            dataCold.append([eval(num.strip('( ')) for num in strs[i][:-1]])

humanFile = "hornHUMAN_1420_906MHzLO_signalRF_maxSamp"
dataHuman =[]
with open(humanFile, 'r') as fh:
    strs = [s.split(')') for s in fh]
    for i in range(len(strs)):
        dataHuman.append([eval(num.strip('( ')) for num in strs[i][:-1]])
        

#once we have the data do: 

deltaT = 300 #T_cold - T_human = 10 Kelvin
voltHuman = np.fft.fft(dataHuman)
voltCold = np.fft.fft(dataCold)

pwrHuman = np.mean(np.real(voltHuman)**2 + np.imag(voltHuman)**2)
pwrCold = np.mean(np.real(voltCold)**2 + np.imag(voltCold)**2)


gain = (deltaT * np.array(pwrCold))/ (np.array(pwrHuman) - np.array(pwrCold))

freqs = np.fft.fftfreq(2048, 1/3.2e6)
intensity = sline * gain 
T_line = np.array(intensity) - np.mean(intensity)

fig, ax = plt.subplots()

ax.plot(np.fft.fftshift(freqs), T_line, color = 'r')
# ax.semilogy(np.fft.fftshift(freqs), np.fft.fftshift(pwrCold), color = 'b', label = 'cold')
# plt.legend()
plt.grid()
plt.title('T Line')
plt.xlabel('frequency [MHz]')
plt.ylabel('Intensity [K]')
plt.xlim()
plt.show()

#y axis is intensity in units of K (temp)

print(gain)

# print(np.mean(gain))
```
